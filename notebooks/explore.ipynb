```python
image_files = []
for root, dirs, files in os.walk("data/processed/train"):
    for file in files:
            image_files.append(os.path.join(root, file))
```

```python
image_file_select = random.sample(image_files, 20)
```

```python
image_file_select
```

```python
fig, axes = plt.subplots(ncols=2,nrows=10, sharex=True, sharey=True)

for i, ax in enumerate(axes.flat):
    ax.imshow(Image.open(image_file_select[i]))
    ax.axis('off')
    ax.set_title(os.path.basename(image_file_select[i]))
plt.show()

```
